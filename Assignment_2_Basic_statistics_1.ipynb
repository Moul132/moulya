```python
import pandas as pd
df = pd.read_csv('sales_data_with_discounts.csv')
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    print(f"Column: {col}")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Mode: {df[col].mode().values[0]:.2f}")
    print(f"Standard Deviation: {df[col].std():.2f}")
    print("---")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('sales_data_with_discounts.csv')

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

fig, axes = plt.subplots(nrows=len(numerical_cols), ncols=1, figsize=(6, 5*len(numerical_cols)))

for i, col in enumerate(numerical_cols):
    df[col].hist(ax=axes[i], bins=40)
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')


fig.tight_layout()


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sales_data_with_discounts.csv')
categorical_col = 'Brand'
freq_dist = df[categorical_col].value_counts()
print("Frequency Distribution:")
print(freq_dist)

prop_dist = df[categorical_col].value_counts(normalize=True)
print("Category Proportions:")
print(prop_dist)

plt.figure(figsize=(8, 6))
sns.countplot(x=categorical_col, data=df)
plt.title("Frequency Distribution of Brand")
plt.xlabel("Brand")
plt.ylabel("Frequency")
plt.show()

corr_matrix = pd.crosstab(df['Brand'], df['Model'])
print("Category Correlations:")
print(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title("Category Correlations between Gender and Occupation")
plt.xlabel("Occupation")
plt.ylabel("Brand")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('sales_data_with_discounts.csv')

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

means = df[numerical_cols].mean()
stds = df[numerical_cols].std()

for col in numerical_cols:
    df[col] = (df[col] - means[col]) / stds[col]

print(df.head())
```
