```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, StandardScaler

df = pd.read_csv('ToyotaCorolla - MLR.csv')


print(df.columns)


print(df.describe())


plt.figure(figsize=(10, 6))
sns.histplot(df['Age_08_04'], kde=True)
plt.title('Age Histogram')
plt.show()

sns.histplot(df['KM'], kde=True)
plt.title('KM Histogram')
plt.show()

sns.histplot(df['HP'], kde=True)
plt.title('HP Histogram')
plt.show()

sns.histplot(df['cc'], kde=True)
plt.title('CC Histogram')
plt.show()

sns.histplot(df['Doors'], kde=True)
plt.title('Doors Histogram')
plt.show()

sns.histplot(df['Weight'], kde=True)
plt.title('Weight Histogram')
plt.show()

sns.histplot(df['Price'], kde=True)
plt.title('Price Histogram')
plt.show()

# Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Age_08_04'])
plt.title('Age Boxplot')
plt.show()

sns.boxplot(x=df['KM'])
plt.title('KM Boxplot')
plt.show()

sns.boxplot(x=df['HP'])
plt.title('HP Boxplot')
plt.show()

sns.boxplot(x=df['cc'])
plt.title('CC Boxplot')
plt.show()

sns.boxplot(x=df['Doors'])
plt.title('Doors Boxplot')
plt.show()

sns.boxplot(x=df['Weight'])
plt.title('Weight Boxplot')
plt.show()

sns.boxplot(x=df['Price'])
plt.title('Price Boxplot')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Age_08_04'], y=df['Price'])
plt.title('Price vs. Age Scatter Plot')
plt.show()

sns.scatterplot(x=df['KM'], y=df['Price'])
plt.title('Price vs. KM Scatter Plot')
plt.show()

sns.scatterplot(x=df['HP'], y=df['Price'])
plt.title('Price vs. HP Scatter Plot')
plt.show()


ohe = OneHotEncoder(sparse_output=False)
fuel_type_encoded = ohe.fit_transform(df[['Fuel_Type']])
fuel_type_encoded = pd.DataFrame(fuel_type_encoded, columns=ohe.get_feature_names_out(['Fuel_Type']))


df.drop('Fuel_Type', axis=1, inplace=True)
df = pd.concat([df, fuel_type_encoded], axis=1)


scaler = StandardScaler()
df[['Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Weight', 'Price']] = scaler.fit_transform(
    df[['Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Weight', 'Price']]
)

print(df.head())
```
